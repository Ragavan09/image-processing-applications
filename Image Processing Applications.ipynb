## Classifier-HaarcaScade Classifier (XML)

```python
import cv2
img = cv2.imread('obama.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier('face.xml')
faces = face_cascade.detectMultiScale(img,1.12)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
plt.figure(figsize=(5,15))
plt.imshow(img)
plt.show()
```

## Eye Detection

```python
img = cv2.imread('obama.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier('face.xml')
eye_cascade = cv2.CascadeClassifier('eye.xml')
faces = face_cascade.detectMultiScale(img,1.12)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    single_face = img[y:y+h,x:x+w]
    eyes = eye_cascade.detectMultiScale(single_face,1.22)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),5)
plt.figure(figsize=(5,15))
plt.imshow(img)
plt.show()
```

##Smile Detection

```python
img = cv2.imread('obama.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier('face.xml')
smile_cascade = cv2.CascadeClassifier('smile.xml')
faces = face_cascade.detectMultiScale(img,1.12)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    single_face = img[y:y+h,x:x+w]
    smiles = smile_cascade.detectMultiScale(single_face,1.25,40)
    for ex,ey,ew,eh in smiles:
        cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),5)
plt.figure(figsize=(5,15))
plt.imshow(img)
plt.show()
```

##live webcam

```python
face_cascade = cv2.CascadeClassifier('face.xml')
eye_cascade = cv2.CascadeClassifier('eye.xml')
cam = cv2.VideoCapture(0)
while True:
    ret,frames = cam.read()
    frames = cv2.flip(frames,1)
    faces = face_cascade.detectMultiScale(frames,1.12)
    for x,y,w,h in faces:
        cv2.rectangle(frames,(x,y),(x+w,y+h),(255,0,0),5)
        single_face = frames[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(single_face,1.22)
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),5)
    cv2.imshow('Webcam',frames)
    if cv2.waitKey(1) == 13:   ## 13 - enter key
        break
cam.release()
cv2.destroyAllWindows()
```

##Object Detection by Color (Blue)

```python
import cv2
import numpy as np
img = cv2.imread('blue.jpg')
back_image = cv2.imread('back.jpg')
img = cv2.resize(img,(800,500))
back_image = cv2.resize(back_image,(800,500))
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
low = np.array([[90,80,80]])
high = np.array([[120,255,255]])
mask_image = cv2.inRange(hsv,low,high)
final_image = cv2.bitwise_and(img,img,mask=mask_image)
result = img - final_image   ## all colours except blue
new = np.where(result==0,back_image,result)
cv2.imshow("My Image",img)
cv2.imshow("HSV Image",hsv)
cv2.imshow("Mask Image",mask_image)
cv2.imshow("Final Image",final_image)
cv2.imshow("Result",result)
cv2.imshow("New Image",new)
cv2.waitKey(0)    ### wait till we press any key
cv2.destroyAllWindows()
```
